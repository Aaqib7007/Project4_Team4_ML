```python
# Scale price data, return, and variance values
df_scaled = StandardScaler().fit_transform(
    df[["ClickThroughRate", "ConversionRate"]]
)
```

```python
market_scaled = pd.DataFrame(
    df_scaled,
    columns=["ClickThroughRate", "ConversionRate"]
)

market_scaled['Ticker'] = df['CampaignChannel']
market_scaled = market_scaled.set_index('Ticker')
market_scaled.head()
```

```python
dummies = pd.get_dummies(df["CampaignChannel"],dtype="int64")
dummies.head()
```

```python
market_scaled = df.copy()  # Ensure df_scaled exists before concatenation
market_scaled = pd.concat([market_scaled, dummies], axis=1)
market_scaled
```

```python
#Step 3: Initialize the K-means model with three clusters and then fit the `market_scaled` DataFrame to the model.
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(market_scaled)
```
