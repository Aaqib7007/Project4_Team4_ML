```python

# Read the CSV file into a Pandas DataFrame
# Set the index using the Ticker column
df_market = pd.read_csv(
    Path("cleaned_data.csv"),
    index_col="CustomerID"
)
# Review the DataFrame

df_market.head()
```

```python
y = df_market['Conversion']

# Separate the X variable, the features
X = df_market.drop(columns=['Conversion','Gender','CampaignChannel','CampaignType',])
```

```python
y.head()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
log_reg_model = LogisticRegression(solver='lbfgs', random_state=1)

# Fit the model using training data
fit_model = log_reg_model.fit(X_train, y_train)
```

```python
# Make a prediction using the testing data
pred = log_reg_model.predict(X_test)
```

```python
# Generate a confusion matrix for the model
test_matrix = confusion_matrix(y_test, pred)
```

```python
# Print the classification report for the model
test_report = classification_report(y_test, pred)
print(test_report)
```
